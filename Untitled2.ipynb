```python
# Extract relevant metadata into a dictionary
dicom_dict = {attr: getattr(dicom_data, attr, None) for attr in dir(dicom_data) if not attr.startswith('_')}

# Load metadata into a pandas DataFrame
df = pd.DataFrame([dicom_dict])

# Inspect the DataFrame
print(df.head())
print(f"DataFrame shape: {df.shape}")
print("DataFrame Info:")
print(df.info())

# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Summary statistics
print("Summary Statistics:")
print(df.describe())
```

```python
# Select numerical columns for visualization
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Visualize distributions
for column in numerical_columns:
    fig, (ax_box, ax_hist) = plt.subplots(1, 2, figsize=(12, 6))

    # Boxplot
    ax_box.boxplot(df[column].dropna(), vert=False, whis=1.5)
    ax_box.set_title(f'{column} - Boxplot')
    ax_box.set_xlabel('Value')

    # Histogram
    sns.histplot(df[column].dropna(), bins=50, kde=True, ax=ax_hist, color='blue')
    ax_hist.set_title(f'{column} - Distribution')
    ax_hist.set_xlabel('Value')
    ax_hist.set_ylabel('Frequency')

    # Adjust layout
    plt.tight_layout()
    plt.show()
```
